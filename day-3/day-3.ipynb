```python
### My backtracking solution was WAY too slow... 

banks = open("input.txt", "r")
#banks = open("lol.txt", "r")

total_joltage = 0

for _bank in banks:
    bank = _bank.strip()
    n = len(bank)
    k = 12  # number of digits to select
    
    dp = {}
    
    def get_max(i, j):
        # base case: no digits needed
        if j == 0:
            return 0
        
        # base case: not enough digits remaining
        if n - i < j:
            return None
        
        # base case: exactly j digits remaining - must take all
        if n - i == j:
            return int(bank[i:])
        
        if (i, j) in dp:
            return dp[(i, j)]
        
        # case where we pick current cell and then go next
        take_rest = get_max(i + 1, j - 1)
        if take_rest is not None:
            take_value = int(bank[i]) * (10 ** (j - 1)) + take_rest
        else:
            take_value = None
        
        # case where we don't pick current cell and go next
        skip_value = get_max(i + 1, j)
        
        result = None
        if take_value is not None and skip_value is not None:
            result = max(take_value, skip_value)
        elif take_value is not None:
            result = take_value
        elif skip_value is not None:
            result = skip_value
        
        dp[(i, j)] = result
        return result
    
    max_num = get_max(0, k)
    total_joltage += max_num
    dp.clear()

print(total_joltage)
```
