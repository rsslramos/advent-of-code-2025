Part One

```python
from collections import deque

input_text = open("example_input.txt", "r")

# grid building
grid = []

for _line in input_text:
    line = _line.strip()
    row = [char for char in line]
    grid.append(row)

# find start
start = None
for i, char in enumerate(grid[0]):
    if char == "S":
        start = (0, i)
        break

# traverse grid as a graph
queue = deque()
seen = set()
total_splits = 0
num_rows, num_cols = len(grid), len(grid[0])

queue.appendleft(start)

while queue:
    size = len(queue)
    num_splits = 0
    
    for i in range(size):
        curr = queue.pop()

        if curr in seen:
            continue
        seen.add(curr)

        curr_row, curr_col = curr

        if curr_row < 0 or curr_row >= num_rows or curr_col < 0 or curr_col >= num_cols:
            continue

        if grid[curr_row][curr_col] == "^":
            queue.appendleft((curr_row, curr_col - 1))
            queue.appendleft((curr_row, curr_col + 1))
            num_splits += 1
        else:
            queue.appendleft((curr_row + 1, curr_col))

    total_splits += num_splits

print(total_splits)
```

```python
input_text = open("input.txt", "r")


# grid building
grid = []

for _line in input_text:
    line = _line.strip()
    row = [char for char in line]
    grid.append(row)

# find start
start = None
for i, char in enumerate(grid[0]):
    if char == "S":
        start = (0, i)
        break

num_rows, num_cols = len(grid), len(grid[0])    
cache = {}

def dp(curr_row, curr_col):
    key = (curr_row, curr_col)
    
    if key == start:
        return 1
    if curr_row < 0 or curr_row >= num_rows or curr_col < 0 or curr_col >= num_cols:
        return 0
    if key in cache:
        return cache[key]
    
    left, right, up = 0, 0, 0
    
    # check if we can come from above
    if curr_row - 1 >= 0 and grid[curr_row - 1][curr_col] != "^":
        up += dp(curr_row - 1, curr_col)
    
    # check for '^' to the left
    if curr_col - 1 >= 0 and grid[curr_row][curr_col - 1] == "^":
        left += dp(curr_row, curr_col - 1)
    
    # check for '^' to the right
    if curr_col + 1 < num_cols and grid[curr_row][curr_col + 1] == "^":
        right += dp(curr_row, curr_col + 1)
    
    cache[key] = left + right + up
    
    return left + right + up

res = sum(dp(num_rows - 1, i) for i in range(num_cols))

print(res)

```
