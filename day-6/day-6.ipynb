Part One

```python
import math 

input_text = open("input.txt", "r")

filtered_rows = []

for _line in input_text:
    line = _line.strip()
    parts = line.split(" ")
    filtered_rows.append([item for item in parts if item != ""])

numbers = filtered_rows[:-1]
operator_set = filtered_rows[-1]

grand_total = 0

num_rows, num_columns = len(numbers), len(numbers[0])

column_groups = []

for i in range(num_columns):
    group = []
    for j in range(num_rows):
        group.append(int(numbers[j][i]))
    column_groups.append(group)
    
grand_total = 0
        
for i, operator in enumerate(operator_set):
    if operator == "*":
        grand_total += math.prod(column_groups[i])
    else:
        grand_total += sum(column_groups[i])
            
print(grand_total)
            
        
```

Part 2

```python
from collections import defaultdict

input_text = open("input.txt", "r")

grid = []


for line in input_text:
    row = []
    for char in line:
        if char != "\n":
            row.append(char)
    grid.append(row)

# omit the last row since it's just operators
num_rows, num_cols = len(grid), len(grid[0])

column_group_map = defaultdict(list)
operator_set = set(["*", "+"])
curr_idx = -1

for i in range(num_cols):
    group = []
    for j in range(num_rows):
        if grid[j][i] != " " and grid[j][i] not in operator_set:
            group.append(grid[j][i])
        if grid[j][i] in operator_set:
            curr_idx += 1
    if group:
        column_group_map[curr_idx].append(int("".join(group)))

operators = [char for char in grid[-1] if char in operator_set]

grand_total = 0

for i, operator in enumerate(operators):
    grand_total = (grand_total + math.prod(column_group_map[i])) if (operator == "*") else (grand_total + sum(column_group_map[i]))

print(grand_total)  
        
```
