Part one

```python
"""
safe has a dial with nly an arrow on it
0 through 99 in order
makes a small click noise as it reaches each number

input is a sequence of rotations, one per line
a rotation starts with an L or R, indicating which direction to turn
followed by a number, the amount to turn

safe starts at position 50

the actual password is number of times the dial is left 
pointing at 0 after any rotation in the sequence.
"""

rotations = open("input.txt", "r")
curr_pos = 50
num_times_at_zero = 0

for rotation in rotations:
    direction = rotation[0]
    amount = int(rotation[1:])
    
    if direction == "R":
        curr_pos = (curr_pos + amount) % 100
    else:
        curr_pos = (curr_pos - amount) % 100
        
    if curr_pos == 0:
        num_times_at_zero += 1

print(num_times_at_zero)
```

Part Two

```python
rotations = open("input.txt", "r")
curr_pos = 50
num_times_at_zero = 0

for rotation in rotations:
    direction = rotation[0]
    amount = int(rotation[1:])

    if direction == "R":
        for i in range(amount):
            curr_pos += 1
            curr_pos %= 100
            if curr_pos == 0:
                num_times_at_zero += 1
    else:
        for i in range(amount):
            curr_pos -= 1
            curr_pos %= 100
            if curr_pos == 0:
                num_times_at_zero += 1

print(num_times_at_zero)
```
