Part 2

```python
ranges_input = open("input.txt", "r")


ranges = ranges_input.read().split(",")
res = 0

seen = set()

for _range in ranges:
    first_id, last_id = _range.split("-")

    
    for i in range(int(first_id), int(last_id) + 1):
        num_as_str = str(i)
        length = len(num_as_str)
        # print(num_as_str)

        if length == 2:
            if i % 11 == 0:
                res += i
            continue

        for j in range(1, length // 2 + 1):

            flag = True

            first_part = num_as_str[:j]
            second_part = num_as_str[j:]
            # print(first_part, second_part)

            if len(second_part) % len(first_part) != 0:
                continue

            l, r = 0, len(first_part)
            while r <= len(second_part):
                if first_part != second_part[l:r]:
                    flag = False
                    break

                l += len(first_part)
                r += len(first_part)

            if flag and i not in seen:
                # print(first_part, second_part, i)
                seen.add(i)
                res += i
    
print(res)
    
```
